```python
#存開盤日
date_mark = []

#匯入json requests 套件
import json
import requests

#將網址get的json資料存在r
stockNumber = "2330"
url = "http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=20170801&stockNo=" + stockNumber
r = requests.get(url)
r.json()
#json.loads(r.text)

#將資料存在data
#使用json.dump() 將資料改為 String
data = r.json()
data_str = json.dumps(data)
#data2 = json.loads(json_str)

#可用 type(檢測變數) 來檢測該變數型態
#data_str
data2 = data['data']
#date = data2[22][0]
#closePrice = data2[22][6]
#["日期","成交股數","成交金額","開盤價","最高價","最低價","收盤價","漲跌價差","成交筆數"]


import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as md
 
    
#迴圈長度為data2的大小
i = 0
X_list = []
Y_list = []
#print 0 to 22
while i<len(data2):
    #print(data2[i][0],data2[i][6])
    ss = data2[i][0]
    ss2 = ss.split('/')
    a = int(ss2[0]) + 1911
    b = int(ss2[1])
    c = int(ss2[2])
    d2 = datetime.date(a,b,c)
    n1 = md.date2num(d2)
    X_list.append(n1)
    
    e = float(data2[i][6])
    Y_list.append(e)
    
    i+=1
    
    date_mark.append(c)
    
    

    
plt.scatter(X_list,Y_list)

plt.title('Stock')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()


from pylab import figure, show
fig = figure()
ax = fig.add_subplot(111)
ax.plot_date(X_list, Y_list, '-')  #將日期及營收傳入繪圖函式
fig.autofmt_xdate()
show()

```
